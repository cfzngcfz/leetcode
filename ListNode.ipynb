# [21. 合并两个有序链表](https://leetcode.cn/problems/merge-two-sorted-lists/)
将两个升序链表合并为一个新的**升序**链表并返回。新链表是通过拼接给定的两个链表的所有节点组成的。

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        # 遍历ListNode
        record = []
        while list1:
            record.append(list1.val)
            list1 = list1.next
        while list2:
            record.append(list2.val)
            list2 = list2.next
        # 排序
        record.sort()
        # 构造ListNode
        if len(record) > 0:
            head2 = ListNode(record[0])
            p = head2
            for ii in range(1,len(record)):
                p.next = ListNode(record[ii])
                p = p.next
            return head2
        else:
            return None

list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(4)

list2 = ListNode(1)
list2.next = ListNode(3)
list2.next.next = ListNode(4)

s = Solution()
output = s.mergeTwoLists(list1, list2)

while output:
    print(output.val)
    output = output.next
```

```python
#include <iostream>
#include <vector>
#include <algorithm>

struct ListNode {
    int val;
    ListNode* next;
    ListNode() : val(0), next(nullptr) {}
    ListNode(int x) : val(x), next(nullptr) {}
    ListNode(int x, ListNode* next) : val(x), next(next) {}
};

class Solution {
public:
    ListNode* mergeTwoLists(ListNode* list1, ListNode* list2) {
        // 遍历ListNode
        std::vector<int> record;
        while (list1) {
            record.emplace_back(list1->val);
            list1 = list1->next;
        }
        while (list2) {
            record.emplace_back(list2->val);
            list2 = list2->next;
        }
        // 排序
        sort(record.begin(), record.end());

        // 构造ListNode
        if (record.size() > 0) {
            ListNode* head2 = new ListNode{ record[0] };
            ListNode* p = head2;
            for (int ii = 1; ii < record.size(); ii++) {
                p->next = new ListNode{ record[ii] };
                p = p->next;
            }
            return head2;
        }
        else {
            return nullptr;
        }
    }
};

int main()
{  
    // 输入
    ListNode* list1 = new ListNode{ 1 };
    list1->next = new ListNode{ 2 };
    list1->next->next = new ListNode{ 4 };
    ListNode* list2 = new ListNode{ 1 };
    list2->next = new ListNode{ 3 };
    list2->next->next = new ListNode{ 4 };

    // 构造类，并执行计算函数
    Solution s;
    ListNode* output = s.mergeTwoLists(list1, list2);
    
    // 打印
    while (output) {
        std::cout << output->val << std::endl;
        output = output->next;
    }

}
```

# [143. 重排链表](https://leetcode.cn/problems/reorder-list/)
给定一个单链表 `L` 的头节点 `head`，单链表 `L` 表示为：  
$L_0 \to L_1 \to \dots \to L_{n - 1} \to L_n$  
请将其重新排列后变为：  
$L_0 \to L_n \to L_1 \to L_{n - 1} \to L_2 \to L_{n - 2} \to \dots$  
不能只是单纯的改变节点内部的值，而是需要实际的进行节点交换。

**示例 2**：  

>**输入**：head = [1,2,3,4,5]  
>**输出**：[1,5,2,4,3]

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)


record = list()
p = head
while p:
    record.append(p)
    p = p.next

left, right = 0, len(record) - 1
while left < right:
    record[left].next = record[right]
    left += 1
    if left == right:
        break
    record[right].next = record[left]
    right -= 1

record[left].next = None


q = head
while q:
    print(q.val)
    q = q.next
```
