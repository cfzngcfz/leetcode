# [3. 无重复字符的最长子串](https://leetcode.cn/problems/longest-substring-without-repeating-characters/)
给定一个字符串`s`，请你找出其中不含有重复字符的**最长***子串*的长度。<br>
*子(字符)串*是字符串中连续的*非空*字符序列。

**示例 1**:
> **输入**: s = "abcabcbb"<br>
**输出**: 3<br> 
**解释**: 因为无重复字符的最长子串是 "abc"，所以其长度为 3。

```python
s = "abcabcbb"

length = 0
left = 0
right = 0
set1 = set()
while right < len(s):
    if s[right] not in set1:
        set1.add(s[right])
        right += 1
    else:
        set1.remove(s[left])
        left += 1
    length = max(length, len(set1))
print(length)
```
