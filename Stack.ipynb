# [20. 有效的括号](https://leetcode.cn/problems/valid-parentheses/)
给定一个只包括`'('`, `')'`, `'{'`, `'}'`, `'['`, `']'`的字符串`s`，判断字符串是否有效。<br>
有效字符串需满足：
1. 左括号必须用相同类型的右括号闭合。
2. 左括号必须以正确的顺序闭合。
3. 每个右括号都有一个对应的相同类型的左括号。


```python
def isValid(s):
    stack = ['?']
    for ss in s:
        if ss == '(' or ss =='[' or ss == '{':
            stack.append(ss)
        elif ss == ')':
            if stack[-1] == '(':
                stack.pop()
            else:
                return False
        elif ss == ']':
            if stack[-1] == '[':
                stack.pop()
            else:
                return False
        else:
            if stack[-1] == '{':
                stack.pop()
            else:
                return False
    return len(stack) == 1

isValid("()[]{}")
```

# [227. 基本计算器 II](https://leetcode.cn/problems/basic-calculator-ii/)
给你一个字符串表达式`s`，请你实现一个基本计算器来计算并返回它的值。<br>
整数除法仅保留整数部分。<br>
你可以假设给定的表达式总是有效的。所有中间结果将在$[-2^{31}, 2^{31} - 1]$的范围内。<br>
注意：不允许使用任何将字符串作为数学表达式计算的内置函数，比如`eval()`。

> **题解**: 每次遍历到数字末尾时，根据`pre_sign`来决定计算方式:
> - 加号：将数字压入栈；
> - 减号：将数字的相反数压入栈；
> - 乘除号：计算数字与栈顶元素，并将栈顶元素替换为计算结果。

```python
s = "14-3/2"

stack = []
pre_sign = '+'
num_str = ""
for ss in s:
    if ss == ' ':
        continue
    elif ss == '+' or ss == '-' or ss == '*' or ss == '/':
        if pre_sign == '+':
            stack.append(int(num_str))
        elif pre_sign == '-':
            stack.append(-int(num_str))
        elif pre_sign == '*':
            stack[-1] *= int(num_str)
        else:
            if stack[-1] >= 0:
                stack[-1] = stack[-1] // int(num_str)
            else:
                stack[-1] = -(-stack[-1] // int(num_str))
        num_str = ""
        pre_sign = ss
    else:
        num_str += ss

if pre_sign == '+':
    stack.append(int(num_str))
elif pre_sign == '-':
    stack.append(-int(num_str))
elif pre_sign == '*':
    stack[-1] *= int(num_str)
else:
    if stack[-1] >= 0:
        stack[-1] = stack[-1] // int(num_str)
    else:
        stack[-1] = -(-stack[-1] // int(num_str))

print(sum(stack))
```

## [224. 基本计算器](https://leetcode.cn/problems/basic-calculator/)
给你一个字符串表达式`s`，请你实现一个基本计算器来计算并返回它的值。<br>
注意:不允许使用任何将字符串作为数学表达式计算的内置函数，比如`eval()`。

> 与[227. 基本计算器 II](#P.227)的区别:
> - 227: `stack`中记录的是 完成乘除运算后的按照加减split的元素, 最后求和
> - 224: `stack`中记录的是 每层括号中的加减运算后的结果, 遇到`(`字符`stack`增加一个元素, 遇到`)`字符`stack`减少一个元素, 同时`pre_sign`也随着`stack`一起变化, 最后返回`stack`中的唯一元素

```python
s = "(1+(4+5+2)-3)+ 6 -(6+8) +9"

stack = [0]
pre_sign = ['+']
num_str = ""
for ss in s:
    if ss == ' ':
        continue
    elif ss == '+' or ss == '-':
        if len(num_str) > 0:
            if pre_sign[-1] == '+':
                stack[-1] += int(num_str)
            else:
                stack[-1] -= int(num_str)
        num_str = ""
        pre_sign[-1] = ss
    elif ss == '(': # stack 与 pre_sign 均增加一维
        stack.append(0)
        pre_sign.append('+')
        num_str = ""
    elif ss == ')':
        # 完成当前()中的加减运算
        if len(num_str) > 0:
            if pre_sign[-1] == '+':
                stack[-1] += int(num_str)
            else:
                stack[-1] -= int(num_str)
        
        # 降维
        val = stack.pop()
        pre_sign.pop()
        
        # 当前()中的结果添加到上一层()中
        if pre_sign[-1] == '+':
            stack[-1] += val
        else:
            stack[-1] -= val
        
        num_str = ""
    else:
        num_str += ss

if len(num_str) > 0:
    if pre_sign[-1] == '+':
        stack[-1] += int(num_str)
    else:
        stack[-1] -= int(num_str)

print(stack[0])
```
